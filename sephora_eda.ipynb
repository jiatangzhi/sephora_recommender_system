## üßº 2. Missing Values Analysis

```python
# Calculate percentage of missing values
missing_percentage = (products_df.isnull().sum() / len(products_df)) * 100

# Sort in descending order
missing_percentage = missing_percentage.sort_values(ascending=False)

# Display rounded percentages
missing_percentage.round(2)
```

In `products_df`, 5 columns have over 40% missing values. Most of them relate to optional pricing and variation details and can be safely dropped or filled with defaults. 


```python
# Drop columns with more than 80% missing values
threshold = 0.8
null_ratio = products_df.isnull().mean()
cols_to_drop = null_ratio[null_ratio > threshold].index
products_df = products_df.drop(columns=cols_to_drop)
```

```python
# Drop high-null optional pricing columns
cols_to_drop_optional = ['child_min_price', 'child_max_price']
products_df = products_df.drop(columns=cols_to_drop_optional, errors='ignore')

# Drop rows with null ratings (if used in modeling or scoring)
products_df = products_df[products_df['rating'].notna()]
```

```python
# Fill common optional fields with 'Unknown' or 'None'
products_df['highlights'] = products_df['highlights'].fillna('None')
products_df['size'] = products_df['size'].fillna('Unknown')
products_df['variation_value'] = products_df['variation_value'].fillna('Unknown')
products_df['variation_type'] = products_df['variation_type'].fillna('Unknown')
products_df['tertiary_category'] = products_df['tertiary_category'].fillna('Uncategorized')
products_df['secondary_category'] = products_df['secondary_category'].fillna('Other')
products_df['ingredients'] = products_df['ingredients'].fillna('No ingredients listed')
```

```python
# Check for remaining missing values
missing = products_df.isnull().sum()
missing_percent = (missing / len(products_df)) * 100
missing_percent[missing_percent > 0].sort_values(ascending=False)
```

```python
# Calculate percentage of missing values
missing_percentage_reviews = (reviews_df.isnull().sum() / len(reviews_df)) * 100

# Sort in descending order
missing_percentage_reviews = missing_percentage_reviews.sort_values(ascending=False)

# Display rounded percentages
missing_percentage_reviews.round(2)
```


In `reviews_df`, all columns are under 40% missing. You can keep everything and fill missing demographic info like skin tone and eye color with "unknown" for modeling or grouping purposes.



```python
# Impute helpfulness with 0
reviews_df['helpfulness'] = reviews_df['helpfulness'].fillna(0)

# Impute missing categorical values with "Unknown"
for col in ['hair_color', 'skin_tone', 'eye_color', 'skin_type']:
    reviews_df[col] = reviews_df[col].fillna('Unknown')

# Impute review_title with placeholder
reviews_df['review_title'] = reviews_df['review_title'].fillna('No Title')

# Impute is_recommended with mode
reviews_df['is_recommended'] = reviews_df['is_recommended'].fillna(reviews_df['is_recommended'].mode()[0])
```

```python
# Check for remaining missing values
missing = reviews_df.isnull().sum()
missing_percent = (missing / len(reviews_df)) * 100
missing_percent[missing_percent > 0].sort_values(ascending=False)
```

### 6.2. Correlation Matrix

```python
# Select only numeric columns of interest
numeric_cols = [
    'rating_x', 
    'price_usd_x',  
    'loves_count',
    'helpfulness',
    'total_feedback_count',
    'total_pos_feedback_count',
    'total_neg_feedback_count'
]

# Drop rows with NaNs in any of these columns (optional)
corr_data = merged_df[numeric_cols].dropna()

# Compute correlation matrix
corr_matrix = corr_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5,
    square=True
)
plt.title('Correlation Matrix of Numeric Variables')
plt.tight_layout()
plt.show()
```

Strong correlations between `total_feedback_count`, `total_pos_feedback_count`, and `total_neg_feedback_count` are expected due to their additive nature. While this multicollinearity could affect regression models, it‚Äôs less critical here since future *collaborative filtering* and *content-based methods* rely on user‚Äìitem interactions or similarity rather than interpreting coefficients. Still, awareness of redundancy helps streamline feature selection if needed later.

### 6.3. Group Differences




```python
# Grouped boxplots by category
group_vars = ['skin_type', 'skin_tone', 'is_recommended']

for var in group_vars:
    plt.figure(figsize=(8, 4))
    
    sns.boxplot(
        data=reviews_df,
        x=var,
        y='rating',
        hue=var,
        palette='Set3',
        dodge=False,
        legend=False
    )
    
    plt.title(f'Rating Distribution by {var.replace("_", " ").title()}')
    plt.xlabel(var.replace("_", " ").title())
    plt.ylabel('Rating')
    plt.xticks(rotation=30, ha='right')
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Step 1: Define mapping from raw values to broader categories
skin_tone_groups = {
    'fair': ['fair', 'light', 'lightMedium', 'fairLight', 'porcelain'],
    'medium': ['medium', 'mediumTan', 'tan', 'olive', 'rich'],
    'dark': ['deep', 'dark', 'ebony'],
    'not sure': ['Not sure', 'notSureST', 'Unknown']
}

# Step 2: Reverse the mapping for `.map()` (flat dictionary)
flat_mapping = {}
for group, values in skin_tone_groups.items():
    for val in values:
        flat_mapping[val] = group

# Step 3: Apply the mapping
reviews_df['skin_tone_grouped'] = reviews_df['skin_tone'].map(flat_mapping).fillna('other')  # handles unexpected values

# Step 4: Check distribution
print(reviews_df['skin_tone_grouped'].value_counts())
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(
    data=reviews_df,
    x='skin_tone_grouped',
    y='rating',
    hue='skin_tone_grouped',      
    palette='pastel',
    dodge=False,
    legend=False                   
)
plt.title('Rating Distribution by Grouped Skin Tone')
plt.xlabel('Grouped Skin Tone')
plt.ylabel('Rating')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### What is the average of `ratings` by product and brand?

##### By product

```python
# Ensure product_id is the same type in both DataFrames
reviews_df['product_id'] = reviews_df['product_id'].astype(str)
products_df['product_id'] = products_df['product_id'].astype(str)

# Calculate the average rating per product
rating_by_product = reviews_df.groupby('product_id')['rating'].mean().reset_index()
rating_by_product.columns = ['product_id', 'avg_rating']

# Merge with product names
rating_by_product = rating_by_product.merge(
    products_df[['product_id', 'product_name']],
    on='product_id',
    how='left'
)

# Display top 10 highest-rated products
top_rated = rating_by_product.sort_values('avg_rating', ascending=False).head(10)
top_rated[['product_name', 'avg_rating']]
```

```python
# Display bottom 10 lowest-rated products
lowest_rated = rating_by_product.sort_values('avg_rating', ascending=True).head(10)
lowest_rated[['product_name', 'avg_rating']]
```

```python
# Filter products with at least 10 reviews
reviews_count = reviews_df.groupby('product_id').size().reset_index(name='review_count')
rating_by_product = rating_by_product.merge(reviews_count, on='product_id')

# Filter products with sufficient number of reviews
filtered = rating_by_product[rating_by_product['review_count'] >= 10]

# Get the bottom 10 lowest-rated products (with enough reviews)
lowest_rated = filtered.sort_values('avg_rating', ascending=True).head(10)

# Display product name, average rating, and number of reviews
lowest_rated[['product_name', 'avg_rating', 'review_count']]
```

##### By brand

```python
# Count number of reviews per product
review_count = reviews_df.groupby('product_id')['rating'].count().reset_index()
review_count.columns = ['product_id', 'num_reviews']

# Average rating per product
avg_rating_per_product = reviews_df.groupby('product_id')['rating'].mean().reset_index()
avg_rating_per_product.columns = ['product_id', 'avg_rating']

# Combine average rating and number of reviews
product_summary = avg_rating_per_product.merge(review_count, on='product_id')

# Merge with product name, brand name, and price
product_summary = product_summary.merge(
    products_df[['product_id', 'product_name', 'brand_name', 'price_usd']],
    on='product_id',
    how='left'
)

# Sort by number of reviews and get top 10
most_reviewed_products = product_summary.sort_values('num_reviews', ascending=False).head(10)

# Display product name, brand, rating, number of reviews, and price
most_reviewed_products[['product_name', 'brand_name', 'avg_rating', 'num_reviews', 'price_usd']]
```
